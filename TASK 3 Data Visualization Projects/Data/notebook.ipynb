### Step 1: Import Libraries And Load Data Set

```python
# import libraries 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("SuperStore_Sales_Dataset.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

### Step 2: Data Cleaning.

```python
# check duplicates values 
df.duplicated().sum()
```

```python
# check null values
df.isnull().sum()
```

```python
# Remove null values
df.dropna(inplace=True)
```

```python
# check null values after removing null values
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower().str.replace(" ", "_")
```

```python
# Convert data types for analysis 
df['order_date'] = pd.to_datetime(df['order_date'])
df['ship_date'] = pd.to_datetime(df['ship_date'])

# convert float to int for analysis
float_cols = df.select_dtypes(include='float').columns
df[float_cols] = df[float_cols].astype(int)

# convert object to int for analysis
df['product_id'] = df['product_id'].astype(int)
df['customer_id'] = df['customer_id'].astype(int)
df['order_id'] = df['order_id'].astype(int)
```

```python
# check data types after conversion
df.info()
```

```python
plt.plot(df['order_date'], df['sales'])
plt.title("Sales Over Time")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x='order_date', y='sales', data=df)
plt.title("Monthly Sales Trend", fontsize=16)
plt.grid(True)
```
